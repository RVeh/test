### Relative Prognoseintervalle mit der Konfidenzellipse

Die Konfidenzellipse erlaubt die Darstellung von Prognoseintervallen,
deren untere und obere Grenzen durch zwei deterministische Funktionen
$f$ und $g$ gegeben sind mit

$$
f(p) = p + z \cdot \sqrt{\frac{p(1-p)}{n}} \quad \text{und} \quad g(p) = p - z \cdot \sqrt{\frac{p(1-p)}{n}},
$$
mit $z = \Phi^{-1}\!\left(\frac{1+\gamma}{2}\right).$

Ausgewertet werden diese Prognoseintervalle für einen bekannten
Parameterwert $p$ (Schnitte parallel zur $y$-Achse). Dabei liegt die Zufälligkeit in der Zufallsgröße
$
H = X/n,
$
deren Realisationen bei einem bevorstehenden Bernoulli-Experiment mit
Wahrscheinlichkeit $\gamma$ im dargestellten Prognoseintervall
(Schnitte auf der $y$-Achse) liegen.


Demgegenüber entstehen Konfidenzintervalle als Realisationen eines
zufälligen Verfahrens: Ihre Lage hängt vom Stichprobenergebnis ab und
entspricht Schnitten parallel zur $x$-Achse.

Die Graphen zu $f$ und $g$ beschreiben somit als feste Grenzfunktionen
die untere bzw. obere Schranke des Prognoseintervalls für $H$, während
Konfidenzintervalle selbst Zufallsintervalle darstellen.


```python
from ci_intervals import (PIConfig, CIStyle,plot_pi)

cfg = PIConfig(
    p=0.63, n=80, gamma=0.95,
    p_min=0.0,  p_max=1.0,
)

style = CIStyle(
    curve_upper="gray", curve_lower="gray",
    interval_bar="tab:blue",
    grid=False,   # False -> kein Gitter, True --> Gitter
    ticks="fine" # fine -> feinere Skalierung, sonst -> "normal"
    )

li,re = plot_pi(
    cfg, 
    style, 
    # Grafik nicht speichern: save=None
    save=f"pi_interval_rel{cfg.p}_n{cfg.n}_gamma{cfg.gamma}.pdf",show_info=True)
```
