### Das Binomialmodell

Wir betrachten das Modell  $X \sim \mathrm{Bin}(n,p)$.

Der Parameter $p$ ist fest,
das Stichprobenergebnis $X$ ist zuf√§llig.

```python
import setup_path
from lib.binomial import (
    BinomialModel,
    DistributionPlotStyle,
    TypographyStyle,
    plot_binomial_distribution,
)

model = BinomialModel(n=100, p=0.45)
style = DistributionPlotStyle(sigma_range=4.0, grid=True)
typo = TypographyStyle()

plot_binomial_distribution(model, style, typo=typo)
```

### Kumulative Wahrscheinlichkeiten

Wir betrachten die Ereignisse
$P(X \le k)$ und $P(X \ge k)$
als aufaddierte Einzelwahrscheinlichkeiten.

```python
import setup_path
from lib.binomial import (
    plot_binomial_distribution_cdf,
    plot_binomial_cdf,
)

plot_binomial_distribution_cdf(
    model,
    style,
    typo,
    k=36,
    direction="le",
)

plot_binomial_cdf(
    model,
    style,
    typo,
    k_mark=35,
    direction="le",
)
```

### Ablehnungsbereiche

Ein Ablehnungsbereich ist eine Setzung
des Testverfahrens.
Er ist keine Entscheidung.

```python
import setup_path
from lib.binomial import (
    BinomialTest,
    plot_binomial_model_with_rejection_region,
)

test = BinomialTest(n=100, p0=0.5, alpha=0.05)

plot_binomial_model_with_rejection_region(
    model=model,
    test=test,
    style=style,
    typo=typo,
    k_obs=45,   # optional
)
```
