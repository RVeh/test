### Prognoseintervalle (relativ und absolut)

Hier wird ein Prognoseintervall für eine Binomialverteilung betrachtet.

- **Relativ:** für $H=X/n$ (Anteil)  
- **Absolut:** für $X$ (Anzahl)

Beide Darstellungen gehören zusammen, haben aber unterschiedliche Achsen und Interpretationen.

### Prognoseintervall als geometrisches Objekt

Im Prognoseintervall ist der Modellparameter $p$ fixiert.  
Der Zufall liegt im Stichprobenergebnis $H = X/n$.

Das Prognoseintervall entsteht als **vertikaler Schnitt** durch die
Konfidenzellipse.  
Die blaue Strecke beschreibt alle möglichen Realisationen,
die unter dem Modell mit Wahrscheinlichkeit $\gamma$ zu erwarten sind.

Optional kann man die zugehörige Stichprobenverteilung
entlang dieser Strecke „mitlesen“:
nicht als neue Grafik, sondern **innerhalb derselben Geometrie**.

So wird sichtbar:
Der Zufall steckt nicht im Parameter $p$, sondern in der Realisation.

### Prognoseintervall und Zufall

Die horizontale Strecke ist das **relative Prognoseintervall**.  
Optional wird die **Stichprobenverteilung** von $H = X/n$ entlang dieser Strecke dargestellt.  
Die Höhen sind **normierte Wahrscheinlichkeiten** – keine zweite Achse.  
So wird sichtbar, **wo der Zufall liegt**: im Stichprobenergebnis.  
Ein zarter horizontaler Schnitt öffnet denselben Raum für das **Konfidenzintervall**.

### Parameter setzen und Grafik ausgeben

```python
# ------- Parameter setzen -------
model = PIModel(
    p=0.4, 
    n=50, 
    gamma=0.95
    )

geometry = IntervalGeometry(
    x_min=0.0, x_max= 1.0
)

style = CIStyle(
    curve_upper="gray", curve_lower="gray",
    area_color="lightgray", area_alpha=0.25,
    interval_color="tab:blue",
    helper_lines="gray",
    grid=False,
)
# ------- Optional: h (Stichprobenergebnis) -------
h = 0.46
# ------- Ausgabe und ggf. Grafik speichern -------
plot_pi(model, geometry, style,
       absolute=False,
       show_distribution = True, dist_scale = 0.8,
       show_ci_slice = True, ci_h = h,
       save=rf"realtiv_PI_n_{model.n}_h_{h}_p_{model.p}.pdf" # save=None --> nicht speichern
       );
```
