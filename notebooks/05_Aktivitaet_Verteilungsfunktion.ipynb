### Verteilungsfunktionen der Binomialverteilung

In dieser Aktivität betrachten wir die **Verteilungsfunktion** der Binomialverteilung.

Die Verteilungsfunktion ist definiert durch
$$ F(k) = P(X \le k). $$

Sie fasst die punktweisen Wahrscheinlichkeiten zu einer **stufigen Funktion** zusammen.
Die Sprunghöhen entsprechen den Einzelwahrscheinlichkeiten $P(X=k)$.

```python
import setup_path

from lib.binomial import (BinomialModel,DistributionPlotStyle,
            TypographyStyle,plot_binomial_cdf,
)
```

```python
model = BinomialModel(n=200, p=0.45)

style = DistributionPlotStyle(
    sigma_range=4.0,
    grid=False,
    figsize=(8, 4),
)

typo = TypographyStyle()

plot_binomial_cdf(model, style, typo,k_mark=80,direction="ge");
```
